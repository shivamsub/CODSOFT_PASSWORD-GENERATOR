```python
#IMPORTING MODULES
import tkinter as tk
import random
import string

#FUNCTION TO GENERATE PASSWORD
def generate_pwd():
    length = int(length_entry.get())
    chars = string.ascii_letters+string.digits+string.punctuation
    password = ''.join(random.choice(chars) for i in range(length))
    generated_pwd_label.config(text=password)

#FUNCTION TO DISPLAY MESSAGE
def display_msg():
    username = username_entry.get()
    password = generated_pwd_label.cget("text")
    message_label.config(text=f"Hello {username}, your generated password is : {password}", fg="red")

#FUNCTION TO RESET
def reset():
    username_entry.delete(0,tk.END)
    length_entry.delete(0,tk.END)
    generated_pwd_label.config(text="")
    message_label.config(text="")

#FUNCTION TO EXIT
def exit_generator():
    window.destroy()

#SETTING UP THE WINDOW
window = tk.Tk()
window.title("PASSWORD GENERATOR")
window.config(bg="light yellow")
window.geometry("300x250")

username_label = tk.Label(window, text="Username : ", bg="light yellow")
username_label.pack()
username_entry = tk.Entry(window)
username_entry.pack()

length_label = tk.Label(window, text="Password Length : ", bg="light yellow")
length_label.pack()
length_entry = tk.Entry(window)
length_entry.pack()

generate_button = tk.Button(window, text="Generate Password", command = generate_pwd)
generate_button.pack()

generated_pwd_label = tk.Label(window, bg="light yellow")
generated_pwd_label.pack()

accept_button = tk.Button(window, text="Accept Password", command = display_msg)
accept_button.pack()

message_label = tk.Label(window, bg="light yellow")
message_label.pack()

reset_button = tk.Button(window, text="Reset", command = reset)
reset_button.pack()

exit_button = tk.Button(window, text="Exit", command = exit_generator)
exit_button.pack()

window.mainloop()
```
